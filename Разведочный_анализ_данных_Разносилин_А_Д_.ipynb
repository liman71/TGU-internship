```python
df['salary_json'] = df['salary'].fillna('{}').apply(eval) # разделим зарплату на "ОТ" и "ДО"
def get_values(json):
  try:
    return json['from']
  except:
    return None
df['salary_from'] = df['salary_json'].apply(get_values)
```

```python
df['salary_json'] = df['salary'].fillna('{}').apply(eval) # колонка с валютой зарплаты
def get_values(json):
  try:
    return json['currency']
  except:
    return None
df['salary_currency'] = df['salary_json'].apply(get_values)
```

```python
rates = dict(RUR=1, USD=53, EUR=56, KZT=0.12) # конвертируем ЗП в иностранной валюте
df['converted_to_rur_min'] = df['salary_currency'].map(rates) * df['salary_from']
df['converted_to_rur_min']
```

```python
rates = dict(RUR=1, USD=53, EUR=56, KZT=0.12) # конвертируем ЗП в иностранной валюте
df['converted_to_rur_max'] = df['salary_currency'].map(rates) * df['salary_to']
df['converted_to_rur_max']
```
